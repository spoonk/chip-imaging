```python
from stitcher.linear_stitcher import LinearStitcher
from imager.imaging_grid import ImagingGrid
import matplotlib.pyplot as plt
```

```python
tiff_path = r"C:\Users\Luke\Desktop\chip-imager\chip-imaging\backend\prototyping\sample_data\test5"
```

```python
grid = ImagingGrid()
# grid.set_properties([0, 0], 20000, 10000, 1800, 0.62)
grid.set_imaging_width(5000)
grid.set_imaging_height(5000)
grid.set_distance_between_images(1300)
grid.set_pixels_per_um(0.62)
```

```python
stitcher = LinearStitcher(tiff_path, grid)
```

```python
images = stitcher._load_tiff_images()
res = stitcher._stitch_images(images)
```

```python
big = stitcher._paste_images_into_canvas(res, images)
```

```python
plt.imshow(big, cmap='gray')
```
